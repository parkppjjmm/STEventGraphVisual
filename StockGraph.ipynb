```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Set the time period
start_date = '2010-01-01'
end_date = '2024-06-25'

# Get Tesla's stock data from Yahoo Finance
tesla_data = yf.download('TSLA', start=start_date, end=end_date)

# Plotting the closing prices of Tesla
plt.figure(figsize=(14, 7))
plt.plot(tesla_data['Close'], label='Tesla Closing Price', color='r')
plt.title('Tesla Stock Price from 2010 to June 2024')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python


import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from adjustText import adjust_text

#   
font_path = 'C:\\Windows\\Fonts\\malgun.ttf'  #    (: Malgun Gothic)
fontprop = fm.FontProperties(fname=font_path, size=9)

#  
start_date = '2020-01-01'
end_date = '2024-06-25'
tesla_data = yf.download('TSLA', start=start_date, end=end_date)

#   
events = {
    '2020-09-22': 'Battery Day',
    '2020-12-21': 'S&P 500  ',
    '2021-03-24': '  ',
    '2021-05-12': '  ',
    '2021-06-10': ' S Plaid ',
    '2021-08-19': 'AI Day',
    '2021-11-11': 'Elon Musk  ',
    '2022-04-25': '  ',
    '2022-10-28': 'Elon Musk Twitter  (440 )',
    '2023-07-01': 'Cybertruck  ',
    '2024-01-30': '  ',
    '2024-06-13': '2024 '
}

#        
events = {pd.to_datetime(date): event for date, event in events.items()}
events = dict(sorted(events.items()))

#      (    )
option_dates = list(events.keys())
call_volumes = [3200, 4500, 3900, 4100, 3700, 3600, 3300, 4200, 4000, 3800, 3400, 3600]
put_volumes = [2900, 4100, 3500, 3900, 3400, 3200, 3100, 4000, 3700, 3500, 3100, 3300]

#  
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.plot(tesla_data['Close'], label='Tesla Closing Price', color='r')

#   
texts = []
for date, event in events.items():
    if date in tesla_data.index:
        event_text = f"{date.strftime('%Y-%m-%d')}\n{event}"
        texts.append(ax1.annotate(event_text, 
                                  (date, tesla_data.loc[date]['Close']), 
                                  xytext=(0, 10), 
                                  textcoords='offset points',
                                  arrowprops=dict(arrowstyle='->', color='blue', lw=1, shrinkA=5, shrinkB=5),
                                  fontsize=9,
                                  fontproperties=fontprop,
                                  horizontalalignment='center',
                                  verticalalignment='bottom'))

#        
ax2 = ax1.twinx()
width = 5  #  
ax2.bar(option_dates, call_volumes, width=width, alpha=0.3, color='g', label='Call Option Volume')
ax2.bar(option_dates, put_volumes, width=width, alpha=0.3, color='b', label='Put Option Volume')

#  
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='blue', lw=1, shrinkA=5, shrinkB=5))

# y   (    )
ax1.set_ylim(tesla_data['Close'].min() * 0.95, tesla_data['Close'].max() * 1.1)

#   
ax1.set_title('2020 2024 6      ', fontproperties=fontprop)
ax1.set_xlabel('', fontproperties=fontprop)
ax1.set_ylabel(' (USD)', fontproperties=fontprop)
ax1.legend(prop=fontprop)
ax1.grid(True)

ax2.set_ylabel(' ', fontproperties=fontprop)
ax2.legend(loc='upper left', prop=fontprop)

plt.show()
```
